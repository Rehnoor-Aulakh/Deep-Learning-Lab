```python
import tensorflow as tf
```

```python
from tensorflow import keras
import matplotlib.pyplot as plt
# plt.style.use("seaborn")
```

```python
# Generate Data Points

X= tf.linspace(-10,10,100)
X
```

```python
X.shape
```

```python
# Sigmoid Activation function

y=tf.keras.activations.sigmoid(X)
# Now let us visualize this

plt.plot(X,y, label = "sigmoid(x)= 1 / (1+exp(-x))")
plt.title("Sigmoid")
plt.xlabel("X")
plt.ylabel("Sigmoid of X")
plt.legend()
plt.show()
```

```python
# tanh

y = tf.keras.activations.tanh(X)
plt.plot(X,y,label = "tanh")
plt.title("Hyperbolic Tangent")
plt.xlabel("X")
plt.ylabel("tanh(X)")
plt.legend()
plt.show()
```

```python
# Relu 

y= tf.keras.activations.relu(X)
plt.plot(X,y,label = "y = max(0,x)")
plt.xlabel("X")
plt.ylabel("max(0,x)")
plt.title("Relu")
plt.legend()
plt.show()
```

```python
# Leaky Relu

y= tf.keras.activations.leaky_relu(X)
plt.plot(X,y)
plt.xlabel("X")
plt.ylabel("Leaky Relu")
plt.title("Leaky Relu")
plt.legend()
plt.show()
```

```python
# Elu

y = tf.keras.activations.elu(X)
plt.plot(X,y,label = "y=x if x>0 else alpha * (exp(x) - 1) if x<0")
plt.title("Elu")
plt.xlabel("X")
plt.ylabel("elu(x)")
plt.legend()
plt.show()
```

```python
# Softmax
x= tf.constant([150,50,10], dtype= tf.float32)
x= tf.expand_dims(x, axis=0)
print(x.shape,x)
y = tf.keras.activations.softmax(x)
print(y)
```
