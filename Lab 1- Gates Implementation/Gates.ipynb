AND Gate

```python
import pandas as pd

df = {
    'x1':[0,0,1,1],
    'x2':[0,1,0,1],
    'y':[0,0,0,1]
}
df= pd.DataFrame(df)
```

```python
df
```

```python
lr = 0.1
w1=1
w2=1
b=0
num_epochs=100
```

```python
for epoch in range(num_epochs):
    for row in df.itertuples():
        x1,x2,y= row.x1,row.x2,row.y
        z= w1*x1 + w2*x2 + b
        y_pred= 1 if z>=0 else 0
        error= y-y_pred
        
        w1+=lr*error*x1
        w2+=lr*error*x2
        b+=lr*error

    
```

```python
print("w1: ",w1)
print("w2: ",w2)
print("b: ",b)
```

```python
#checking
for x1,x2 in [(0,0),(0,1),(1,0),(1,1)]:
    z=w1*x1 + w2*x2 + b
    y_pred=1 if z>=0 else 0
    print(f"x1={x1}, x2={x2} => y_pred={y_pred}")
```

Single Layer Perceptron is giving wrong output

```python
import math
import random
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+math.exp(-x))
def sigmoid_derivative(x):
    return x*(1-x)
```

```python
X=[
    [0,0],
    [0,1],
    [1,0],
    [1,1]
]
Y=[0,1,1,0]
```

```python
w_h = [[random.uniform(-1,1),random.uniform(-1,1)] for _ in range(2)]
b_h = [random.uniform(-1,1) for _ in range(2)]

w_o = [random.uniform(-1,1), random.uniform(-1,1)]
b_o = random.uniform(-1,1)

lr=0.5
```

```python
for epoch in range(10000):
    total_error=0
    #4 data points, set the weights accordingly
    for i in range(4):
        x1,x2= X[i]
        y= Y[i]
        
        #---------------------Forward pass---------------------
        h=[]
        for j in range(2):
            z= w_h[j][0]*x1 + w_h[j][1]*x2 +b_h[j]
            h.append(sigmoid(z)) 
        z_out = w_o[0]*h[0] + w_o[1]*h[1] + b_o
        output= sigmoid(z_out)
        
        #--------------------Error --------------------
        error= y-output
        total_error+=error**2
        
        #----------------------Backpropagation----------------
        d_out = error*sigmoid_derivative(output)
        
        #Update output weights
        for j in range(2):
            w_o[j]+=lr*d_out*h[j]
        b_o+=lr*d_out
        
        #Hidden layer updates
        for j in range(2):
            d_h = d_out*w_o[j]*sigmoid_derivative(h[j])
            w_h[j][0]+=lr*d_h*x1
            w_h[j][1]+=lr*d_h*x2
            b_h[j]+=lr*d_h

        if epoch % 1000 == 0:
            print(f"Epoch {epoch}, Error {total_error:.4f}")

            
        
```

```python
print("\nFinal XOR predictions:")
for x in X:
    h1 = sigmoid(w_h[0][0]*x[0] + w_h[0][1]*x[1] + b_h[0])
    h2 = sigmoid(w_h[1][0]*x[0] + w_h[1][1]*x[1] + b_h[1])
    out = sigmoid(w_o[0]*h1 + w_o[1]*h2 + b_o)
    print(x, round(out))
```
