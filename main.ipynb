Part B: Define the model

```python
def forward(X,w,b):
    # print(X.shape)
    # print(w.shape)
    y_hat=np.dot(X,w)+b
    # print(y_hat.shape)
    return y_hat

#Checkpoint:
#parameters are: length, diameter, shell weight
#number of parameters: 3
```

Part C: Define Loss (MSE)

```python
def mse(y,y_hat):
    #y is a list, and y_hat is a list
    return np.sum((y-y_hat)**2)/len(y)

# Checkpoint:

# why square:
# because square will consider only the magnitude of error, and assign higher penalty to big error and smaller penalty to smaller error

# what mistakes are expensive:
# mistakes with a larger error are expensive    
```

Part D: The Learning Rule

```python
#update the weights based on gradients
#what gradient means in words:
#gradient is the slope, which means how much and in what direction is the loss affected when the weight is changed

#why subtracting gradient reduces loss:
#if gradient is positive, means the loss is increasing with the increase in weight, so we need to subtract the gradient to reduce the loss
#if gradient is negative, means that loss is decreasing with increase in weight, so we need to add the gradient
```

```python
def grad_w(X,y,y_hat):
    #TODO: return dW
    n=len(y)
    return -2 * np.dot(X.T, (y-y_hat))/n
    
def grad_b(X,y,y_hat):
    #TODO: return db
    n=len(y)
    return -2 * np.sum(y-y_hat)/n
```

```python
#Checkpoint
#meaning of large gradient:
#large gradient means that the slope is very large, a large effect in loss will take place after updating the weight

#effect of too-large learning rate:
#If the learning rate is very large, then the weight might overshoot the decision space and go on the other side, which will have an opposite effect to loss.
```

Part E: Training loop

```python
#initialize w (small random values)
w= np.random.uniform(low=-1,high=1,size=3)

#initialize b (zero)
b=0

#choose learning rate and epochs
lr=0.01
epochs=1000
```

```python
for epoch in range(epochs):
    #1. forward pass
    y_hat = forward(X_train_normalized,w,b)
    
    #2. compute loss
    loss= mse(y_train,y_hat)
    
    #3. compute gradients
    w_grad= grad_w(X_train_normalized,y_train,y_hat)
    b_grad= grad_b(X_train_normalized,y_train,y_hat)
    
    #4. update w and b 
    w= w-(lr*w_grad)
    b= b-(lr*b_grad)
    
    if epoch%10==0:
        print(epoch,loss)
```

```python
#Initial expectation: loss would decrease fast 
#Revised expectation after training: loss first decreases fast and then decreases slowly
```

```python
# Compute test MSE
y_test_pred = forward(X_test_normalized, w, b)
test_mse = np.sum((y_test - y_test_pred)**2) / len(y_test)
print(f"Test MSE: {test_mse}")
```

```python
# Compute test MAE
test_mae = np.sum(np.abs(y_test - y_test_pred)) / len(y_test)
print(f"Test MAE: {test_mae}")
```
