Task 1.1- Input Generation

```python
import numpy as np
import pandas as pd
```

```python
X= np.linspace(0,10,50).reshape(-1,1)
X
```

```python
X.shape
```

```python
#X represents a linearly growing straightline from 0 to 10 which has 50 points
```

```python
noise = np.random.normal(0,0.2, size=(50,1))
print(noise)
```

```python
y = np.log(X+1) + noise
print(y)
```

```python
W1= np.random.uniform(-0.5,0.5, size=(1,3))
b1= np.ones((1,3))*0.1

W2= np.random.uniform(-0.5,0.5, size=(3,1))
b2= np.ones((1,1))*0.1

```

```python
W1
```

```python
W1.shape
```

```python
def activation(z):
    return np.maximum(0,z)
```

```python
def activation_slope(z):
    return (z>0).astype(float)
```

```python
z1= X@W1 + b1
```

```python
z1
```

```python
h= activation(z1)
h
```

```python
y_hat = h@W2+b2
```

```python
y_hat
```

Define Error and Loss

```python
error = y_hat-y
```

```python
loss= np.mean(error**2)
loss
```

Backward Flow

```python
dL_dy = 2* error/len(X)
```

```python
# gradient of loss w.r.t W2 and b2 (output layer)
# dL/dW2 = h.T @ dL_dy
dL_dW2 = h.T @ dL_dy
dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)
print(dL_dW2.shape, dL_db2.shape)
```

```python
# gradient flowing back into hidden layer
# dL/dh = dL/dy_hat * dy_hat/dh = dL_dy @ W2.T
dL_dh = dL_dy @ W2.T
print(dL_dh.shape)
```

```python
# apply relu slope to get gradient w.r.t z1
# where relu is 0, gradient is blocked (that neuron didn't contribute)
dL_dz1 = dL_dh * activation_slope(z1)
print(dL_dz1.shape)
```

```python
# gradients for first layer W1 and b1
dL_dW1 = X.T @ dL_dz1
dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)
print(dL_dW1.shape, dL_db1.shape)
```

Parameter Update (Step 9)

```python
# move parameters opposite to the gradient to reduce loss
# same idea as perceptron: w = w - lr * slope
learning_rate = 0.01

W1 -= learning_rate * dL_dW1
b1 -= learning_rate * dL_db1
W2 -= learning_rate * dL_dW2
b2 -= learning_rate * dL_db2

print("weights updated")
```

Training Loop (Step 10)

```python
# re-initialize weights fresh before training
np.random.seed(42)
W1 = np.random.uniform(-1, 1, size=(1, 3))
b1 = np.zeros((1, 3))
W2 = np.random.uniform(-1, 1, size=(3, 1))
b2 = np.zeros((1, 1))

learning_rate = 0.01
epochs = 2000
loss_history = []

for epoch in range(epochs):
    # forward pass
    z1 = X @ W1 + b1
    h = activation(z1)
    y_hat = h @ W2 + b2

    # loss
    error = y_hat - y
    loss = np.mean(error ** 2)
    loss_history.append(loss)

    # backward pass
    dL_dy = 2 * error / len(X)

    dL_dW2 = h.T @ dL_dy
    dL_db2 = np.sum(dL_dy, axis=0, keepdims=True)

    dL_dh = dL_dy @ W2.T
    dL_dz1 = dL_dh * activation_slope(z1)

    dL_dW1 = X.T @ dL_dz1
    dL_db1 = np.sum(dL_dz1, axis=0, keepdims=True)

    # update parameters
    W1 -= learning_rate * dL_dW1
    b1 -= learning_rate * dL_db1
    W2 -= learning_rate * dL_dW2
    b2 -= learning_rate * dL_db2

    if epoch % 200 == 0:
        print(f'epoch {epoch}  loss = {loss:.5f}')

print(f'final loss = {loss:.5f}')
```

```python
import matplotlib.pyplot as plt

# plot loss over epochs to see if training worked
plt.plot(loss_history)
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.show()
```

```python
# final prediction vs actual
z1_final = X @ W1 + b1
h_final = activation(z1_final)
y_pred = h_final @ W2 + b2

plt.scatter(X, y, label='actual (noisy)', s=15)
plt.plot(X, np.log(X + 1), label='true curve', color='green', linestyle='--')
plt.plot(X, y_pred, label='model prediction', color='red')
plt.title('Model Prediction vs Actual')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
