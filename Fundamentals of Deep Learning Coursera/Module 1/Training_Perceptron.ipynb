```python
import numpy as np
import pandas as pd
```

```python
class Perceptron:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.lr = learning_rate
        self.n_iters= num_iterations
        self.activation_function= self._relu_function
        self.weights = None
        self.bias= None
        
    def _relu_function(self, x):
        return np.maximum(0,x)

    def fit(self, X,y):
        n_samples, n_features= X.shape
        
        #initialize the weights and bias
        
        self.weights= np.zeros(n_features)
        self.bias= 0
        
        for _ in range(self.n_iters):
            for index , x_i in enumerate(X):
                linear_output= np.dot(x_i, self.weights)+ self.bias
                y_predicted= self.activation_function(linear_output)
                
                # Perceptron Update Rule
                # In Real life scenerios, gradient rule will be used instead of direct error calculation
                update = self.lr * (y[index] - y_predicted)
                self.weights += update * x_i
                self.bias += update
                
    def predict(self, X):
        linear_output= np.dot(X, self.weights)+ self.bias
        y_predicted= self.activation_function(linear_output)
        return y_predicted
        
        
```

```python
X= np.array([[0,0],[0,1],[1,0],[1,1]]) #Logical AND Dataset
y= np.array([0,0,0,1])
```

```python
perceptron = Perceptron(learning_rate=0.1, num_iterations=100)
perceptron.fit(X,y)
```

```python
predictions = perceptron.predict(X)
print("Predictions: ", predictions)
```
